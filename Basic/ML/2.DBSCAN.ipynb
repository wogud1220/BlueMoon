```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
```

```python
#임의의 점 군집을 만든다.
#centers = center가 될(군집의 중심이 될) 3개의 포인트
centers = [[1,1], [-1, -1], [0,0]]

#X는 데이터셋으로 사용할 X의 위치.
#n_samples=포인트의 수, centers=중심점의 위치, cluster_std=얼마나 흩어져있을 것인지.
X, label = make_blobs(n_samples=300, centers=centers, cluster_std=0.5)
#랜덤 데이터이기 때문에 사용하기 위해서 스케일링 적용
X = StandardScaler().fit_transform(X)
```

```python
#scatter(산점도 X, Y)
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
#print(X)
```

```python
db = DBSCAN(eps=0.4, min_samples=10).fit(X)
```

```python
label = db.labels_
```

```python
print(label, type(label))
```
