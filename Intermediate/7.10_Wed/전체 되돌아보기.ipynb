```python
#필요 없는 열 정리
df.drop(['id', 'Unnamed: 32'], axis='columns', inplace= True)
```

```python
df.diagnosis.replace({'M':1, 'B':0}, inplace= True)
# object -> int 변환
df
```

```python
df.corrwith(df.diagnosis).plot(kind='bar', grid=True, figsize=(12,10), title="Correlation of dignosis")

# diagnosis 열과 다른 열과의 상관관계를 바그래프로 만듦
```

```python
x = [1,2,3,4,5]
y =[3,4,2,6,7]

# 시각화 그래프를 2행 1열로 표현
# figsize = 도화지 사이즈
# ax = 도화지를 접어서 그릴지, 그대로 그릴지 2행 1열로 접겠다.
fig, ax = plt.subplots(2,1, figsize=(12,6))
# linestyle = 선의 종류
ax[0].plot(x, y, c='r', linestyle='--', marker ='*')
# c 는 색상이름 ok, 몇 개의 생상이 필요한지도 list 형태로 가능,  5개라서 5개의 색상
ax[1].scatter(x, y, c=y)
plt.show()
```

```python
# train_test_split
from sklearn.model_selection import train_test_split
X = df.drop(['diagnosis'], axis=1) # 진단 빼고 전부 다
Y = df['diagnosis']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
df
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
# 모델링
svc_cls = SVC()
svc_cls.fit(X_train, Y_train)

pred_svc = svc_cls.predict(X_test)
accuracy_score(Y_test, pred_svc)
```

```python
from sklearn.model_selection import GridSearchCV
# 그리드 리서치 
# 사이키릿런 svm. svc에 가서 parameters 검색하여 찾기
parameters = { 'kernel' : ('poly', 'sigmoid'),
              'gamma' : ('scale', 'auto') 
}
svc_p = SVC()
svc_g = GridSearchCV(svc_p, parameters)
svc_g.fit(X_train, Y_train)

print(f'Best parameter : {svc_g.best_params_}')
print(f'Best scores : {svc_g.best_score_}')
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_cls = KNeighborsClassifier(n_neighbors=5, metric = 'minkowski')
knn_cls.fit(X_train, Y_train)
pred_knn = knn_cls.predict(X_test)
accuracy = accuracy_score(Y_test, pred_knn)
accuracy
```

```python
# 랜덤 서치
from sklearn.model_selection import RandomizedSearchCV
knn_r = KNeighborsClassifier()
parameters = {"n_neighbors" : [1, 3, 9, 10]}
knn_cv = RandomizedSearchCV(knn_r, parameters)
knn_cv.fit(X_train, Y_train)

print(f'Best parameter : {knn_cv.best_params_}')
print(f'Best scores : {knn_cv.best_score_}')

```

```python
from xgboost import XGBClassifier
xgb_cls = XGBClassifier()
xgb_cls.fit(X_train, Y_train)

#프레딕트
pred_xgb = xgb_cls.predict(X_test)
accuracy = accuracy_score(pred_xgb,X_test)


```
