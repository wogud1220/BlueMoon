```python
import pandas as pd
df = pd.read_csv('./git.csv')
df
```

```python
df.shape
```

```python
df.describe()
```

```python
#cabin, age 결측치 확인
df.info()
```

```python
#평균값 대체
d_mean = df['Age'].mean()
d_mean
```

```python
df['Age'].fillna(d_mean, inplace=True)
```

```python
d_mode = df['Embarked'].mode()[0]
df['Embarked'].fillna(d_mode,inplace=True)
```

```python
#cabin 삭제
df.drop(columns='Cabin', inplace=True)
df.head()
```

```python
df.info()
```

```python
#원 핫 인코딩
from sklearn.preprocessing import LabelEncoder

#Sex, Embarked, 각각 2개, 3개로 나눌 수 있음
# M, F를 0과 1로 변환
le = LabelEncoder()
df['en_sex'] = le.fit_transform(df['Sex'])
df['en_embarked'] = le.fit_transform(df['Embarked'])
```

```python
df
```

```python
X = df[['Pclass', 'en_sex', 'en_embarked', 'Age']]
Y = df['Survived']
```

```python
#분석 데이터 분할

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test  = train_test_split(X, Y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
# 의사 결정 트리
dt = DecisionTreeClassifier()
dt.fit(X_train, Y_train)
y_predict = dt.predict(X_test)
from sklearn.metrics import accuracy_score
acc = accuracy_score(Y_test,y_predict)
print(acc)
```

```python
#분류르르 다양한 모델로 적용해보기
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# 랜덤 포레스트 사용
X_train, X_test, Y_train, Y_test  = train_test_split(X, Y, test_size=0.2)
dr = RandomForestClassifier(n_estimators=20, max_depth=3)
# 20그루의 나무로 3번의 분기
dr.fit(X_train, Y_train)
y_predict = dr.predict(X_test)
acc = accuracy_score(Y_test,y_predict)
print(acc)
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# knn사용

X_train, X_test, Y_train, Y_test  = train_test_split(X, Y, test_size=0.2)
dk = KNeighborsClassifier(n_neighbors=5)
dk.fit(X_train, Y_train)
y_predict = dk.predict(X_test)
acc = accuracy_score(Y_test,y_predict)
print(acc)
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# svm 사용

X_train, X_test, Y_train, Y_test  = train_test_split(X, Y, test_size=0.2)



ds = svm.SVC(kernel='rbf')
'''
kernel='rbf'는 Support Vector Machine (SVM)에서 사용할 커널 함수의 종류를 지정하는 파라미터입니다. SVM은 데이터를 고차원 공간으로 변환하여 분류 경계를 최대한으로 나누는 초평면을 찾습니다. 이때 커널 함수는 데이터를 고차원 공간으로 변환하는 방법을 정의합니다.

rbf는 Radial Basis Function의 약자로, SVM에서 자주 사용되는 커널 함수 중 하나입니다. RBF 커널은 가우시안 커널이라고도 불리며, 비선형 데이터를 분류하는 데 유용합니다.

RBF 커널의 특징
비선형 변환:

RBF 커널은 데이터를 고차원 공간으로 변환하여 비선형 경계를 찾을 수 있게 합니다. 이는 선형 커널로는 분류할 수 없는 데이터셋을 분류하는 데 효과적입니다.
'''
ds.fit(X_train, Y_train)
y_predict = ds.predict(X_test)
acc = accuracy_score(Y_test,y_predict)
print(acc)
```
