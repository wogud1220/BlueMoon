```python
# 3. 데이터의 전처리
#결측치가 전체의 30% 이상인 경우 해당 열을 제거하시오. 400의 30퍼 280개 이하면 제거
#결측치가 전체의 30% 미만인 경우, 평균값으로 대체하시오. 나머진느 평균으로 대체
# d_mean = df['en_htn'].mean()
# d_mean
# df['en_htn'].fillna(d_mean, inplace=True)

d_mode = df['ane'].mode()[0]
df['ane'].fillna(d_mode,inplace=True)
```

```python
df.info()
```

```python
# 데이터의 전처리 2

#원 핫 인코딩
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['en_pe'] = le.fit_transform(df['pe'])
```

```python
X = df[['en_htn', 'en_ane', 'en_pcv', 'en_wc',
       'en_pc', 'en_ba', 'en_dm', 'en_cad', 'en_appet', 'en_pe','age', 'bp', 'sg', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo']]
Y = df['classification']
```

```python
#분석 데이터 분할

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test  = train_test_split(X, Y, test_size=0.3)
```

```python
#의사 결정 트리
from sklearn.tree import DecisionTreeClassifier
# 의사 결정 트리
dt = DecisionTreeClassifier()
dt.fit(X_train, Y_train)
y_predict = dt.predict(X_test)
from sklearn.metrics import accuracy_score
acc = accuracy_score(Y_test,y_predict)
print(acc)
```

```python
#분류르르 다양한 모델로 적용해보기
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# 랜덤 포레스트 사용
X_train, X_test, Y_train, Y_test  = train_test_split(X, Y, test_size=0.2)
dr = RandomForestClassifier(n_estimators=20, max_depth=3)
# 20그루의 나무로 3번의 분기
dr.fit(X_train, Y_train)
y_predict = dr.predict(X_test)
acc = accuracy_score(Y_test,y_predict)
print(acc)
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# knn사용

X_train, X_test, Y_train, Y_test  = train_test_split(X, Y, test_size=0.2)
dk = KNeighborsClassifier(n_neighbors=5)
dk.fit(X_train, Y_train)
y_predict = dk.predict(X_test)
acc = accuracy_score(Y_test,y_predict)
print(acc)
```
